```python
import numpy as np
import pandas as pd
import geopandas as gpd

import folium
import seaborn as sns
```

```python
sf = gpd.read_file("../data/tables/SA2_2021_AUST_GDA2020.shp")
sf.head()
```

```python
_map = folium.Map(location=[40.7793, -73.974], tiles="Stamen Terrain", zoom_start=9)

geoJSON = sf[['SA2_CODE21', 'geometry']].drop_duplicates("SA2_CODE21").to_json()

# refer to the folium documentations on how to plot aggregated data.
_map.add_child(folium.Choropleth(
    geo_data=geoJSON,
    name='choropleth',
))

_map.save('../plots/general_sa2_visuals.html')
_map
```
